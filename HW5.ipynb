# 1. Когда используется критерий Стьюдента, а когда Z –критерий?
Z-Критерий используется, когда известна дисперсия генеральной совокупности. Критерий Стьюдента (t-критерий) - когда дисперсия генеральной совокупности неизвестна

# 2. Проведите тест гипотезы. Утверждается, что шарики для подшипников, изготовленные автоматическим станком, имеют средний диаметр 17 мм. Используя односторонний критерий с α=0,05, проверить эту гипотезу, если в выборке из n=100 шариков средний диаметр оказался равным 17.5 мм, а дисперсия известна и равна 4 кв. мм.

```python
mean_gen = 17
alpha = 0.05
n = 100
mean_n = 17.5
var_gen = 4

std_gen = 4 ** (1/2)
z_stat = (mean_n - mean_gen) / (std_gen / n ** (1/2))

crit_value = 1.645
print(f'Z статистика выборки ({z_stat}) больше ее критического значения при статистической значимости 5% ({crit_value}),') 
print('благодаря чему мы можем отвергнуть нулевую гипотезу в пользу альтернативной')
```

# 3. Проведите тест гипотезы. Продавец утверждает, что средний вес пачки печенья составляет 200 г. Из партии извлечена выборка из 10 пачек. Вес каждой пачки составляет: 202, 203, 199, 197, 195, 201, 200, 204, 194, 190. Известно, что их веса распределены нормально. Верно ли утверждение продавца, если учитывать, что доверительная вероятность равна 99%? (Провести двусторонний тест.)

```python
mean_gen = 200
n = 10
weights = [202, 203, 199, 197, 195, 201, 200, 204, 194, 190]
alpha = 0.01

mean_n = np.mean(weights)
std_n = math.sqrt(sum([(weight - mean_n) ** 2 for weight in weights]) / (n - 1))

t_stat = (mean_gen - mean_n) / (std_n / n ** (1/2))
crit_value = 3.69

print(f't-статистика ({round(t_stat, 2)}) меньше, чем критическое значение при двустороннем уровне значимости {alpha / 2} ({crit_value}),')
print('что не дает нам достаточно доказательств для отвержения нулевой гипотезы')
```

# 4. Есть ли статистически значимые различия в росте дочерей? Рост матерей 172, 177, 158, 170, 178,175, 164, 160, 169, 165 Рост взрослых дочерей: 173, 175, 162, 174, 175, 168, 155, 170, 160

```python
heights_1 = [172, 177, 158, 170, 178, 175, 164, 160, 169, 165]
heights_2 = [173, 175, 162, 174, 175, 168, 155, 170, 160]
alpha = 0.025

t_stat, p_value = stats.ttest_ind(heights_1, heights_2, equal_var=False)

if p_value <= 1 - alpha and p_value >= alpha:
    print('Статистически значимых различий нет')
else:
    print('Статистически значимые различия имеются')
```

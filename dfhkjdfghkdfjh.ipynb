```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

def fill_ax(X, Y, ax):
    model = LinearRegression()
    X = X.reshape((len(X), 1))
    model.fit(X, Y)

    R = model.score(X, Y)
    slope = model.coef_.item()
    intercept = model.intercept_

    Y_pred = intercept + slope * X

    ax.scatter(X, Y)
    ax.plot(X, Y_pred, color='red')
    ax.set_title(f'R^2 = {R:.2f}, slope = {slope:.2f}, intercept = {intercept:.2f}');
```

```python
fig, axs = plt.subplots(2, 2, figsize=(10, 10))

X1= np.array([30,30,40, 40])
Y1= np.array([37, 47, 50, 60])
fill_ax(X1, Y1, axs[0, 0])

x2= np.array([30,30,40, 40, 20, 20, 50, 50])
y2= np.array([37, 47, 50, 60, 25, 35, 62, 72])
fill_ax(x2, y2, axs[0, 1])

X3 = np.array([30,30,40, 40, 20, 20, 50, 50, 10, 10, 60, 60])
Y3 = np.array([37, 47, 50, 60, 25, 35, 62, 72, 13, 23, 74, 84]) 
fill_ax(X3, Y3, axs[1, 0])

axs[1, 1].axis('off');
```
